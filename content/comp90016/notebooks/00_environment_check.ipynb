# COMP90016 JupyterLite Environment Check

Use this notebook to confirm what is available in the browser runtime.

JupyterLite can run many Python libraries, but native command-line genomics tools from the old Binder environment are not available in-browser.

```python
import importlib
import platform
import sys

print(f'Python: {sys.version}')
print(f'Platform: {platform.platform()}')
print(f'Pyodide runtime: {"pyodide" in sys.modules}')
```

```python
packages_to_test = [
    ('numpy', 'numpy'),
    ('pandas', 'pandas'),
    ('matplotlib', 'matplotlib'),
    ('seaborn', 'seaborn'),
    ('Bio', 'biopython'),
    ('plotly', 'plotly'),
    ('bqplot', 'bqplot'),
    ('itables', 'itables'),
    ('fastcluster', 'fastcluster'),
    ('pysam', 'pysam'),
]

results = []
for module_name, label in packages_to_test:
    try:
        module = importlib.import_module(module_name)
        version = getattr(module, '__version__', 'unknown')
        results.append((label, 'OK', version))
    except Exception as exc:
        results.append((label, 'NOT AVAILABLE', str(exc)))

for label, status, detail in results:
    print(f'{label:12} {status:14} {detail}')
```

```python
unsupported_native_tools = [
    'blast',
    'bwa',
    'fastqc',
    'fastqe',
    'iqtree',
    'mlst',
    'prokka',
    'samtools',
    'snp-dists',
]

print('Native CLI tools that are not available inside JupyterLite:')
for tool in unsupported_native_tools:
    print(f' - {tool}')

print('\nUse Binder or a local conda environment for those workflows.')
```
