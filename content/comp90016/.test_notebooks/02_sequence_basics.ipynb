# COMP90016 Sequence Basics in the Browser

A small example based on public workshop data.

```python
from pathlib import Path
from Bio import SeqIO
import pandas as pd

fasta = Path('..') / 'data' / 'workshop' / '07' / 'SARS_CoV_2_S.fasta'
records = list(SeqIO.parse(str(fasta), 'fasta'))
print('Records:', len(records))
```

```python
def gc_fraction(seq):
    seq = str(seq).upper()
    gc = seq.count('G') + seq.count('C')
    return gc / len(seq) if seq else 0.0

rows = []
for rec in records:
    rows.append({
        'id': rec.id,
        'length': len(rec.seq),
        'gc_fraction': gc_fraction(rec.seq),
    })

summary = pd.DataFrame(rows)
summary
```

```python
import matplotlib.pyplot as plt

ax = summary['gc_fraction'].plot(kind='hist', bins=20, figsize=(6, 4), title='GC Fraction Distribution')
ax.set_xlabel('GC Fraction')
plt.show()
```
