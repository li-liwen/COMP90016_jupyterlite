# COMP90016 Data Loading Basics

This notebook demonstrates loading public COMP90016 data files from the JupyterLite filesystem.

```python
from pathlib import Path

base = Path('..') / 'data'
print('Data root:', base.resolve() if base.exists() else base)

for top in ['assignment', 'exam', 'workshop']:
    p = base / top
    count = sum(1 for _ in p.rglob('*') if _.is_file())
    print(f'{top}: {count} files')
```

```python
import pandas as pd

isolates_path = base / 'exam' / 'exam_b' / 'data' / 'comp90016_exam_B_isolates.csv'
isolates = pd.read_csv(isolates_path)
isolates.head()
```

```python
from Bio import SeqIO

fasta_path = base / 'workshop' / '07' / 'SARS_CoV_2_S.fasta'
record = next(SeqIO.parse(str(fasta_path), 'fasta'))
print('Sequence ID:', record.id)
print('Length:', len(record.seq))
print('First 80 nt:', str(record.seq)[:80])
```
